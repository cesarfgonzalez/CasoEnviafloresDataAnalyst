```python
df=pd.read_csv('assesment_04_v2_msg.csv',encoding="ISO-8859-1")
df.head()
```

```python
ClasificacionCumpleaños = ['feliz', 'cumpleanos', 'cumple', 'regalo','detalle','sorpresa','felicidades']
ClasificacionPareja = ['amo', 'amor', 'miamor', 'corazon','aniversario','mivida','vida','contigo','lado','esposa','esposo','novio','novia']
ClasificacionGraduacionPromo = ['ingeniero','ingenieria','licenciado','licenciada','lic','arq','arqui','arquitecto','arquitecta','orgulloso','orgullosa','orgullo','exito']
ClasificacionEnfermedadDuelo = ['siento','cielo','paz','descansa','descanse','superar','superas','enfermedades','enfermo','enferma','superaras','superaremos']
ClasificacionMadres = ['ma','mama','mom','mommi','momy','mother','mami','mamita']
ClasificacionAbuelos =['abuelo','abuela','aguelo','aguela','guelo','guela','titi','tito','tita','abuelos']
ClasificacionPerdon = ['perdon','perdob','losiento','perdoname','perdonas','perdonar']
ClasificacionVacio = ['nan']
```

```python
df['Cumpleaños']=0
```

```python
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        for k in range(len(ClasificacionCumpleaños)):
            if ClasificacionCumpleaños[k-1] == df.iloc[i,j]:
                df.loc[i,'Cumpleaños']=df.loc[i,'Cumpleaños']+1

df

            
```
